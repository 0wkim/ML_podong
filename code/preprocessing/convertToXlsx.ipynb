### Convert xls to xlsx

kobis 데이터를 받았을 때 xls 형식으로 다운로드 되는데, 
해당 확장자의 파일을 바로 열게 되면 오류가 발생하여 xlsx로 변환하는 과정이 필요합니다.

그래서 이 파일은 xls 확장자를 xlsx로 변환하는 코드입니다.

1년 기준으로 5~10분 이내에 실행됩니다.

```python
import os
import time
import pythoncom
from win32com.client import DispatchEx

# 1) COM 초기화
pythoncom.CoInitialize()

# 2) Excel 실행
excel = DispatchEx("Excel.Application")
excel.Visible = False
excel.DisplayAlerts = False

# 3) 경로 설정
BASE_DIR   = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
raw_dir    = os.path.join(BASE_DIR, "data", "raw(xls)")
pre_dir    = os.path.join(BASE_DIR, "data", "raw(xlsx)")
os.makedirs(pre_dir, exist_ok=True)  # 폴더 없으면 생성

# 4) raw 폴더의 모든 .xls 파일 처리
for fname in os.listdir(raw_dir):
    if not fname.lower().endswith(".xls"):
        continue

    raw_path = os.path.join(raw_dir, fname)
    print(f"Converting: {raw_path}")

    # (1) 파일 열기
    wb = excel.Workbooks.Open(
        Filename=raw_path,
        ReadOnly=False,
        IgnoreReadOnlyRecommended=True,
        Local=True
    )
    time.sleep(0.5)  # Excel 준비 대기

    # (2) 저장 경로(.xlsx)
    base, _      = os.path.splitext(fname)
    new_fname    = base + ".xlsx"
    new_path     = os.path.join(pre_dir, new_fname)

    # (3) 저장
    wb.SaveAs(new_path, FileFormat=51)
    time.sleep(0.2)

    # (4) 닫기
    wb.Close(SaveChanges=False)

# 5) Excel 종료
excel.Quit()
print("All done!")
```
