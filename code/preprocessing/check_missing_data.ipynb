```python
import os
import pandas as pd

# CSV 파일이 들어있는 폴더 경로
folder_path = '../../data/'

# 불러올 CSV 파일 이름
csv_filename = 'data.csv'

# 전체 경로 생성
csv_path = os.path.join(folder_path, csv_filename)

# pandas로 CSV를 DataFrame으로 읽어들이기
df = pd.read_csv(csv_path)
```

```python
df_unique = df.drop_duplicates(subset=['영화명'], keep='first').reset_index(drop=True)
```

```python
df_unique
```

```python
keep_cols = [
    '영화명', '조회일',
    '매출액증감', '관객수증감(전일대비)',
    '매출액', '관객수',
    '스크린수', '상영횟수',
    '대표국적', '국적',
    '제작사', '배급사',
    '등급', '장르',
    '감독', '배우'
]

df_unique = df_unique[keep_cols].copy()
```

```python
df_unique
```

```python
# 컬럼별 결측값 개수 구하기
missing_counts = df_unique.isnull().sum()

# 결측값이 하나라도 있는 행의 개수 구하기
num_rows_with_na = df_unique.isnull().any(axis=1).sum()

# 결과 출력
print("컬럼별 결측값 개수:")
print(missing_counts)

print(f"\n결측값이 하나라도 있는 행 개수: {num_rows_with_na}")
```
