```python
import os
import pandas as pd

# CSV 파일이 들어있는 폴더 경로
folder_path = '../../data/raw/cultureRaw'

# 불러올 CSV 파일 이름
csv_filename = 'merged_culture_data.csv'

# 전체 경로 생성
csv_path = os.path.join(folder_path, csv_filename)

# pandas로 CSV를 DataFrame으로 읽어들이기
df_culture = pd.read_csv(csv_path)
```

```python
import os
import pandas as pd

# CSV 파일이 들어있는 폴더 경로
folder_path = '../../data/raw/screenAndShowRaw'

# 불러올 CSV 파일 이름
csv_filename = '2_df_screen.csv'

# 전체 경로 생성
csv_path = os.path.join(folder_path, csv_filename)

# pandas로 CSV를 DataFrame으로 읽어들이기
df_screen = pd.read_csv(csv_path)
```

```python
import os
import pandas as pd

# CSV 파일이 들어있는 폴더 경로
folder_path = '../../data/raw/screenAndShowRaw'

# 불러올 CSV 파일 이름
csv_filename = '2_df_showing.csv'

# 전체 경로 생성
csv_path = os.path.join(folder_path, csv_filename)

# pandas로 CSV를 DataFrame으로 읽어들이기
df_showing = pd.read_csv(csv_path)
```

```python
import pandas as pd

# df_screen에서 필요한 컬럼만 미리 추출
screen_cols = ['영화명', '날짜', '스크린점유율', '누적매출액', '누적관객수']
df_screen_sub = df_screen[screen_cols]

# 두 데이터프레임을 영화명·조회일(=날짜) 기준으로 left join
df_merged = pd.merge(
    df_culture,
    df_screen_sub,
    how='left',
    left_on=['영화명', '조회일'],
    right_on=['영화명', '날짜']
)

# '날짜' 컬럼은 중복이므로 제거
df_merged = df_merged.drop(columns=['날짜'])

# 결과 확인
print(df_merged.head())
```

```python
import pandas as pd

# df_merged: 이전에 df_culture와 df_screen을 합친 결과
# df_showing: '영화명', '날짜', '상영점유율' 컬럼을 가진 데이터프레임

# df_showing에서 필요한 컬럼만 추출
showing_cols = ['영화명', '날짜', '상영점유율']
# 1) '날짜'를 '조회일'로 이름 변경
df_showing_sub = df_showing[['영화명', '날짜', '상영점유율']] \
                    .rename(columns={'날짜': '조회일'})

# 2) 공통 키('영화명','조회일')로 left join
df_final = pd.merge(
    df_merged,
    df_showing_sub,
    on=['영화명', '조회일'],
    how='left'
)
```

```python
df_final
```

```python
df_final.to_csv(r'../../data/data.csv', index=False, encoding='utf-8-sig')
```

```python
import os
import pandas as pd

# CSV 파일이 들어있는 폴더 경로
folder_path = '../../data/raw/mergedRaw'

# 불러올 CSV 파일 이름
csv_filename = '2_filled_merged_all.csv'

# 전체 경로 생성
csv_path = os.path.join(folder_path, csv_filename)

# pandas로 CSV를 DataFrame으로 읽어들이기
df_inform = pd.read_csv(csv_path)
```

```python
import pandas as pd

# 1) 두 DF에서 중복되는 컬럼(키 컬럼 제외)을 찾아 제거하기
key_cols = ['영화명', '조회일']
common_cols = [col for col in df_inform.columns 
               if col in df_final.columns and col not in key_cols]

df_inform_sub = df_inform.drop(columns=common_cols)

# 2) '영화명'과 '조회일'을 키로 left join
df_joined = pd.merge(
    df_final,
    df_inform_sub,
    on=key_cols,
    how='left'
)
```
