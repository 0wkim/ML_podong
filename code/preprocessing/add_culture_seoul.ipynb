```python
import re
import pandas as pd

# 정규표현식: Timestamp('…') 안의 날짜 문자열 캡처
pattern = r"Timestamp\('([^']+)'\)"

def parse_date_list(val):
    # 문자열이 아니면 그대로 반환
    if not isinstance(val, str):
        return val
    # 'YYYY-MM-DD HH:MM:SS' 형태의 문자열만 추출
    date_strs = re.findall(pattern, val)
    # pandas.Timestamp → 'YYYY-MM-DD' 문자열로 변환
    return [pd.to_datetime(s).strftime('%Y-%m-%d') for s in date_strs]

# 적용
df['조회일'] = df['조회일'].apply(parse_date_list)

# 확인
print(df['조회일'].iloc[0])          # -> ['2020-10-28', ...]
print(type(df['조회일'].iloc[0][0])) # -> <class 'str'>
```

```python
import pandas as pd

def get_year_month(date_str):
    """
    'YYYY-MM-DD' 형태의 문자열을 받아서 (연도, 월) 튜플로 반환.
    """
    dt = pd.to_datetime(date_str)
    return dt.year, dt.month

# ── 3. 메트릭 컬럼 리스트 정의 ─────────────────────────────────────────────
metrics = [
    '상영영화수',
    '영화관람객수',
    '지출금액',
    '인구당영화관람객수',
    '영화관람객수점수',
    '지출금액점수',
    '종합점수',
    '기준상영영화수',
    '기준영화관람객수',
    '기준지출금액'
]

# ── 4. 조회일(날짜 리스트)을 인자로 받아서, 각 날짜별로 filtered_df에서 메트릭 값을 추출해 딕셔너리로 반환 ───────
def extract_metrics_for_dates(date_list, filtered_df):
    # 결과를 담을 딕셔너리 초기화
    result = {metric: [] for metric in metrics}

    for date_str in date_list:
        year, month = get_year_month(date_str)

        # filtered_df에서 해당 연도, 월을 가진 행 찾기
        matched = filtered_df[
            (filtered_df['BASE_YEAR'] == year) &
            (filtered_df['BASE_MT'] == month)
        ]

        if not matched.empty:
            # 첫 번째(유일하다고 가정) 행에서 각 metric 값을 꺼내어 리스트에 추가
            row = matched.iloc[0]
            for metric in metrics:
                result[metric].append(row[metric])
        else:
            # 해당 연월이 없으면 None 또는 NaN을 추가
            for metric in metrics:
                result[metric].append(None)

    return result

# ── 5. df에 새로운 컬럼(각 메트릭 + "_리스트" 형태) 추가 ──────────────────────────
for metric in metrics:
    df[f"{metric}"] = df['조회일'].apply(
        lambda date_list: extract_metrics_for_dates(date_list, filtered_df)[metric]
    )
```

```python
# 1) metrics 리스트가 이전과 동일하게 정의되어 있어야 합니다.
metrics = [
    '상영영화수',
    '영화관람객수',
    '지출금액',
    '인구당영화관람객수',
    '영화관람객수점수',
    '지출금액점수',
    '종합점수',
    '기준상영영화수',
    '기준영화관람객수',
    '기준지출금액'
]

# 2) 각 metric+"_리스트" → metric 으로 매핑 딕셔너리 생성
rename_map = {f"{metric}_리스트": metric for metric in metrics}

# 3) rename 수행 (inplace=True 로 원본 df 컬럼명 변경)
df.rename(columns=rename_map, inplace=True)

# 4) 최종 컬럼명 확인
print(df.columns.tolist())
```

```python
# 1) metrics 리스트를 그대로 사용
metrics = [
    '상영영화수',
    '영화관람객수',
    '지출금액',
    '인구당영화관람객수',
    '영화관람객수점수',
    '지출금액점수',
    '종합점수',
    '기준상영영화수',
    '기준영화관람객수',
    '기준지출금액'
]

# 2) 'metric' → '서울_metric' 형태로 매핑 딕셔너리 생성
rename_map = {metric: f"서울_{metric}" for metric in metrics}

# 3) df.rename() 으로 한 번에 컬럼명 변경 (inplace=True 로 원본 변경)
df.rename(columns=rename_map, inplace=True)

# 4) 최종 컬럼명 확인
print(df.columns.tolist())
```
