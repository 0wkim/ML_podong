```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 날짜 변환
df_last7['조회일'] = pd.to_datetime(df_last7['조회일'])

# % 제거 및 숫자로 변환
df_last7['스크린점유율'] = (
    df_last7['스크린점유율']
    .astype(str)
    .str.replace('%', '', regex=False)
    .astype(float)
)

# 정렬 후 전일 대비 변화량 계산
df_last7 = df_last7.sort_values(['영화명', '조회일'])

# ✅ 이상치 기준 (전체 기준, 상하 0.5%)
q_low = df_last7['스크린점유율(전일대비)'].quantile(0.001)
q_high = df_last7['스크린점유율(전일대비)'].quantile(0.999)

# ✅ 이상치만 추출
outliers = df_last7[
    (df_last7['스크린점유율(전일대비)'] < q_low) |
    (df_last7['스크린점유율(전일대비)'] > q_high)
]

plt.figure(figsize=(14, 7))

# 고유 영화 수만큼 컬러 생성
titles = outliers['영화명'].unique()
colors = cm.get_cmap('tab20', len(titles))

for idx, title in enumerate(titles):
    # 전체 7일 데이터
    full = df_last7[df_last7['영화명'] == title].sort_values('조회일').tail(7).reset_index(drop=True)
    outlier_vals = outliers[outliers['영화명'] == title]

    x = range(1, len(full) + 1)
    y = full['스크린점유율(전일대비)']

    color = colors(idx)

    # 선 그래프
    plt.plot(x, y, marker='o', color=color)

    # 이상치 표시 (붉은 점)
    for i, row in full.iterrows():
        if row['조회일'] in outlier_vals['조회일'].values:
            plt.scatter(i + 1, row['스크린점유율(전일대비)'], color='red', zorder=5)

    # 영화명 텍스트로 직접 표시 (마지막 지점 기준)
    plt.text(7.1, y.iloc[-1], title, fontsize=9, color=color, verticalalignment='center')

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.title('이상치 영화의 최근 7일 스크린점유율(전일대비) 추이 (x=7이 최신)')
plt.xlabel('최근 7일 순서 (1 = 가장 오래된 날, 7 = 최신)')
plt.ylabel('스크린점유율(전일대비)')
plt.xticks(range(1, 8))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 날짜 정리
df_last30['조회일'] = pd.to_datetime(df_last30['조회일'])

# 전체 사용
df_filtered = df_last30.copy()

# 영화별 정렬 및 패딩
aligned_data = []

for title, group in df_filtered.groupby('영화명'):
    group_sorted = group.sort_values('조회일', ascending=False).head(30)
    group_sorted = group_sorted.sort_values('조회일').reset_index(drop=True)

    missing = 30 - len(group_sorted)
    if missing > 0:
        group_sorted = pd.concat([
            pd.DataFrame({'관객수': [None]*missing}),
            group_sorted[['관객수']]
        ], ignore_index=True)

    aligned_data.append(group_sorted['관객수'].tolist())

# 평균 관객수 계산
aligned_df = pd.DataFrame(aligned_data)
mean_values = aligned_df.mean(skipna=True)

# ✅ 1차 미분
diff1 = np.diff(mean_values)

# ✅ 2차 미분
diff2 = np.diff(diff1)

# ✅ 2차 미분 그래프
plt.figure(figsize=(12, 6))
plt.plot(range(3, 31), diff2, marker='o', color='green', label='2차 미분: 변화량의 변화 (가속도)')

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.title('최근 30일 평균 관객수의 변화량 변화 (2차 미분)')
plt.xlabel('최근 30일 순서 (3 = 세 번째 날, 30 = 최신)')
plt.ylabel('변화량의 변화 (가속도)')
plt.axhline(0, color='gray', linestyle='--')
plt.xticks(range(3, 31))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
