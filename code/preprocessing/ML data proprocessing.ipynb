```python
import pandas as pd

def merge_duplicate_movies_include_all(df):
    movie_col = '영화명'
    count_cols = ['매출액증감', '관객수증감(전일대비)']
    keep_cols = ['개봉일', '대표국적', '국적', '배급사', '등급', '장르', '감독']
    list_cols = ['매출액', '관객수', '스크린수', '상영횟수', '조회일']
    mean_cols = ['매출액(점유율)']
    last_cols = ['누적매출액', '누적관객수']

    # 1. 존재하는 컬럼만 필터링
    count_cols = [col for col in count_cols if col in df.columns]
    keep_cols = [col for col in keep_cols if col in df.columns]
    list_cols = [col for col in list_cols if col in df.columns]
    mean_cols = [col for col in mean_cols if col in df.columns]
    last_cols = [col for col in last_cols if col in df.columns]

    # 2. 조회일을 datetime으로 변환 (정렬을 위해)
    if '조회일' in df.columns:
        df['조회일'] = pd.to_datetime(df['조회일'], errors='coerce')

    # 3. 중복 영화명 추출
    counts = df[movie_col].value_counts()
    dup_titles = counts[counts > 1].index

    # 4. 중복 영화만 필터링 & 조회일 기준 최신순 정렬
    df_dup = df[df[movie_col].isin(dup_titles)].sort_values(by=[movie_col, '조회일'], ascending=[True, False])

    # 5. 병합 방식 정의
    agg_dict = {col: lambda x: list(x) for col in count_cols + list_cols}
    agg_dict.update({col: 'first' for col in keep_cols})
    agg_dict.update({col: 'mean' for col in mean_cols})
    agg_dict.update({col: 'last' for col in last_cols})

    # 6. groupby로 병합
    df_merged = df_dup.groupby(movie_col).agg(agg_dict).reset_index()
    df_merged['중복횟수'] = df_dup[movie_col].value_counts()[df_merged[movie_col]].values

    # 7. 중복 아닌 영화는 그대로 유지
    df_unique = df[~df[movie_col].isin(dup_titles)].copy()
    df_unique['중복횟수'] = 1

    # 8. 병합 + 단일영화 결합
    df_final = pd.concat([df_merged, df_unique], ignore_index=True)

    return df_final
```
